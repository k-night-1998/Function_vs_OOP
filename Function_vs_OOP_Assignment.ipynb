```python
class Solution:
    def sort012(self, arr):
        low = 0
        mid = 0
        high = len(arr) - 1

        while mid <= high:
            if arr[mid] == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1
            elif arr[mid] == 1:
                mid += 1
            else:  # arr[mid] == 2
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1
        return arr

# Example usage
sol = Solution()
print(sol.sort012([0, 1, 2, 0, 1, 2]))
```

```python
class Solution:
    def binarySearch(self, arr, k):
        left = 0
        right = len(arr) - 1
        result = -1

        while left <= right:
            mid = left + (right - left) // 2
            if arr[mid] == k:
                result = mid
                right = mid - 1  # look for earlier occurrence
            elif arr[mid] < k:
                left = mid + 1
            else:
                right = mid - 1
        return result

# Example usage
sol = Solution()
print(sol.binarySearch([1, 1, 1, 1, 2], 1))
print(sol.binarySearch([11, 22, 33, 44, 55], 445))
```
